```python
import pandas as pd
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
dir(diabetes)
```

```python
diabetes.feature_names, diabetes.data_filename, diabetes.target[:5], diabetes.target_filename
```

```python
df = pd.DataFrame(diabetes.data , columns = diabetes.feature_names )
df.head()
```

```python
df['target'] = diabetes.target
df.head(150)
```

```python
x = df.drop('target', axis = 'columns')
x.head()
```

```python
df['target'] = df['target'].replace(
    to_replace=[df[df.target >= 100], df[df.target <= 100]], 
    value=['1', '0'])
df.head()
```

```python
y = df.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.80)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

logmodel = LogisticRegression(solver = 'liblinear', max_iter =300)
linmodel = LinearRegression()
treemodel = DecisionTreeClassifier()
svmmodel = SVC(kernel='linear')
rfmodel = RandomForestClassifier(n_estimators = 10)

logmodel.fit(x_train,y_train)
linmodel.fit(x_train,y_train)
treemodel.fit(x_train,y_train)
svmmodel.fit(x_train,y_train)
rfmodel.fit(x_train,y_train)

print('logistic model score: ', logmodel.score(x_test,y_test))
print('linear model score: ', linmodel.score(x_test,y_test))
print('Decisiontree model score: ', treemodel.score(x_test,y_test))
print('svm model score: ', svmmodel.score(x_test,y_test))
print('random forest model score: ', rfmodel.score(x_test,y_test))
```
